```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
 
#导入数据
df = pd.read_csv('test.csv')
print(df.dtypes)
```

```python
# 折线图
# SOC 变化曲线，速度变化
# 数据预处理
df2=df[['uploadtime','vehiclespeed','soc']]
print(df2.shape)

import matplotlib.pyplot as plt


fig,ax1 = plt.subplots(figsize=(15, 4))

ax2 = ax1.twinx()#产生一个ax1的镜面坐标

ax1.plot(df2['vehiclespeed'],"g-")
plt.grid()
ax1.set_ylabel("vehiclespeed[km/h]")
ax1.set_xlabel("time")
ax1.set_ylim(0, 120)
ax1.set_xlim(0, 6500)

ax2.plot(df2['soc'],"b-")
ax2.set_ylabel("SOC[%]")
ax1.set_ylim(0, 100)
plt.legend({'vehiclespeed','soc'}) 


plt.show()
```

```python
# 使用Seaborn画折线图
df2=df[['uploadtime','vehiclespeed','soc']]
sns.lineplot( x='vehiclespeed', data=df2)
plt.show()
```

```python
# 散点图，电机转矩与转速分布图
#数据预处理，去除零点
df1=df[['emspeed','emtq']]
print(df1.shape)
df1 = df1[ (df1['emspeed']!=0) & (df1['emtq']!=0)]
print(df1.shape)
```

```python
# matplotlib scatter
fig = plt.figure()  
ax1 = fig.add_subplot(111)
plt.grid()
ax1.set_title('EM Working point')  #设置标题  
plt.xlabel('EM_speed')  #设置X轴标签  
plt.ylabel('EM_torque')  #设置Y轴标签  
ax1.scatter(df1['emspeed'], df1['emtq'],c = 'b',marker = '.')  #画散点图  
plt.legend('x1')  #设置图标  
plt.show()  #显示所画的图 
```

```python
sns.set_style("darkgrid") 
sns.jointplot(x='emspeed', y='emtq', #设置xy轴
                data = df1,  #设置数据
                color = 'b', #设置颜色
                s = 10, linewidth = 1,#设置散点大小、及宽度(只针对scatter), #edgecolor = 'w'边缘颜色 #stat_func=sci.pearsonr,计算两个变量之间的相关系数
                kind = 'scatter',#设置类型：'scatter','reg','resid','kde','hex'#stat_func=<function pearsonr>,
                space = 0.1, #设置散点图和布局图的间距
                height = 8, #图表大小(自动调整为正方形))
                ratio = 5, #散点图与布局图高度比，整型
                marginal_kws = dict(bins=15, rug =True), #设置柱状图箱数，是否设置rug
                )
```

```python
sns.set_style("whitegrid") 
sns.jointplot(x='emspeed', y='emtq', #设置xy轴，显示columns名称
                data = df1,  #设置数据
                color = 'b', #设置颜色b
                #stat_func=sci.pearsonr,计算两个变量之间的相关系数
                kind = 'hex',#设置类型：'scatter','reg','resid','kde','hex'
                space = 0.1, #设置散点图和布局图的间距
                height = 8, #图表大小(自动调整为正方形))
                ratio = 2, #散点图与布局图高度比，整型
                )
```

```python
sns.set_style("dark") 
g = sns.JointGrid(x ='emspeed', y='emtq', data=df1)
g = g.plot_joint(sns.kdeplot, cmap = 'Reds_r')     #绘制密度图
plt.grid(linestyle = '--')
g.plot_marginals(sns.kdeplot, shade = True, color = 'r') #绘制x,y轴密度图
```

```python
sns.set(style="dark", palette="muted", color_codes=True) 
g = sns.jointplot(x ='emspeed', y='emtq', data=df1,
                  kind = 'kde', color = 'g', shade_lowest = False)
#添加散点图
g.plot_joint(plt.scatter, c = 'w', s = 3, linewidth = 0.1, marker='+')
```
